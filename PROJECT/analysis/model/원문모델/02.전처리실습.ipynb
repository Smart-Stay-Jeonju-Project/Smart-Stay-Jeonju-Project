```python
# 중복 제거
train_data.drop_duplicates(subset=['sentence'], inplace=True)
test_data.drop_duplicates(subset=['sentence'], inplace=True)
```

```python
import numpy as np
# 결측치 제거
train_data['sentence'] = train_data['sentence'].str.replace('[^가-힣0-9 ]','',regex=True)
train_data['sentence'] = train_data['sentence'].str.replace('^ +','',regex=True)
train_data.dropna(how='any', inplace=True)
```

```python
test_data['sentence'] = test_data['sentence'].str.replace('[^가-힣0-9 ]','',regex=True)
test_data['sentence'] = test_data['sentence'].str.replace('^ +','',regex=True)
test_data.dropna(how='any', inplace=True)
```

```python
# 형태소 분석기 객체 생성
from konlpy.tag import Okt, Kkma, Komoran, Hannanum
okt  = Okt()
```

```python
# 작업의 진행 상황을 시각적으로 표시해주는 라이브러리
from tqdm import tqdm
```

```python
X_train = []
for sentence in tqdm(train_data['sentence'][:1000]) :
    #print(f"처리할 문장 : {sentence}")
    # 토큰화
    #tokenized_sentence = kkma.morphs(sentence)
    #tokenized_sentence = kmrn.morphs(sentence)
    #tokenized_sentence = hnnm.morphs(sentence)
    tokenized_sentence = okt.morphs(sentence)
    # 불용어 제거
    filtered_tokens = [
        word
        for word in tokenized_sentence
        if not word in stopwords
    ]
    X_train.append(filtered_tokens)
```

```python
X_test = []
for sentence in tqdm(test_data['sentence'][:1000]) :
    tokenized_sentence = okt.morphs(sentence)
    # 불용어 제거
    filtered_tokens = [
        word
        for word in tokenized_sentence
        if not word in stopwords
    ]
    X_test.append(filtered_tokens)
```
